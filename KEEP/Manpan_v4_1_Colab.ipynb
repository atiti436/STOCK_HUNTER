```python
# ==================== é…ç½®åƒæ•¸ ====================
CONFIG = {
    # å®ˆè­·è€… 1ï¼šåƒ¹å€¼æŠ•è³‡ï¼ˆå·´è²ç‰¹ï¼‰
    "VALUE_MIN_PE": 5,
    "VALUE_MAX_PE": 20,
    "VALUE_MIN_ROE": 0.15,
    "VALUE_MIN_PROFIT_MARGIN": 0.10,
    "VALUE_MIN_DIV_YIELD": 0.03,
    
    # å®ˆè­·è€… 2ï¼šæˆé•·è‚¡ï¼ˆå½¼å¾—æ—å€ï¼‰
    "GROWTH_MIN_REVENUE_GROWTH": 0.15,
    "GROWTH_MAX_PEG": 1.5,
    "GROWTH_MIN_PROFIT_GROWTH": 0.20,
    
    # å®ˆè­·è€… 3ï¼šæŠ€è¡“åˆ†æï¼ˆå¨å»‰æ­å°¼çˆ¾ï¼‰
    "TECH_MIN_RS_RANK": 70,
    "TECH_MIN_VOLUME_RATIO": 1.5,
    
    # å®ˆè­·è€… 4ï¼šé¢¨éšªç®¡ç†ï¼ˆé›·é”é‡Œæ­ï¼‰
    "RISK_MAX_DEBT_RATIO": 0.60,
    "RISK_MIN_CURRENT_RATIO": 1.2,
    "RISK_MAX_BETA": 1.5,
    
    # å®ˆè­·è€… 5ï¼šå¸‚å ´æƒ…ç·’ï¼ˆç´¢ç¾…æ–¯ï¼‰
    "SENTIMENT_MIN_INST_BUY": 0.60,
    "SENTIMENT_MAX_SHORT_RATIO": 0.05,
    
    # å®ˆè­·è€… 6ï¼šç¸½é«”ç¶“æ¿Ÿï¼ˆé”é‡Œæ­ï¼‰
    "MACRO_MIN_GDP_GROWTH": 0.02,
    "MACRO_MAX_INTEREST_RATE": 0.05,
}

print("âœ… åƒæ•¸è¨­å®šå®Œæˆ")
```

## ğŸ“¦ ç¬¬äºŒæ­¥ï¼šè¼‰å…¥ä¸»ç¨‹å¼

```python
import random
import json
from datetime import datetime, timedelta

# ==================== Mock Data ç”Ÿæˆå™¨ ====================
def generate_mock_stocks(num=50, market_trend="bull"):
    """ç”Ÿæˆæ¨¡æ“¬è‚¡ç¥¨è³‡æ–™"""
    stocks = []
    base_multiplier = 1.2 if market_trend == "bull" else 0.8
    
    for i in range(num):
        ticker = f"STOCK{i+1:03d}"
        
        # åŸºç¤æ•¸æ“š
        price = random.uniform(10, 500) * base_multiplier
        pe = random.uniform(5, 50)
        roe = random.uniform(0.05, 0.30)
        
        stock = {
            # åŸºæœ¬è³‡è¨Š
            "ticker": ticker,
            "name": f"å…¬å¸ {ticker}",
            "price": round(price, 2),
            
            # å®ˆè­·è€… 1ï¼šåƒ¹å€¼æŠ•è³‡
            "pe_ratio": round(pe, 2),
            "roe": round(roe, 4),
            "profit_margin": round(random.uniform(0.05, 0.25), 4),
            "dividend_yield": round(random.uniform(0, 0.08), 4),
            
            # å®ˆè­·è€… 2ï¼šæˆé•·è‚¡
            "revenue_growth": round(random.uniform(-0.10, 0.40), 4),
            "eps_growth": round(random.uniform(-0.20, 0.50), 4),
            "peg_ratio": round(random.uniform(0.5, 3.0), 2),
            
            # å®ˆè­·è€… 3ï¼šæŠ€è¡“åˆ†æ
            "rs_rank": random.randint(1, 99),
            "volume_ratio": round(random.uniform(0.5, 3.0), 2),
            "price_above_ma50": random.choice([True, False]),
            
            # å®ˆè­·è€… 4ï¼šé¢¨éšªç®¡ç†
            "debt_ratio": round(random.uniform(0.10, 0.80), 4),
            "current_ratio": round(random.uniform(0.8, 2.5), 2),
            "beta": round(random.uniform(0.5, 2.0), 2),
            
            # å®ˆè­·è€… 5ï¼šå¸‚å ´æƒ…ç·’
            "institutional_buy_ratio": round(random.uniform(0.30, 0.90), 4),
            "short_interest_ratio": round(random.uniform(0, 0.15), 4),
            "analyst_rating": round(random.uniform(1, 5), 1),
            
            # å®ˆè­·è€… 6ï¼šç¸½é«”ç¶“æ¿Ÿ
            "sector": random.choice(["ç§‘æŠ€", "é‡‘è", "é†«ç™‚", "æ¶ˆè²»", "å·¥æ¥­"]),
            "market_cap": random.randint(100, 10000) * 1000000,
        }
        stocks.append(stock)
    
    return stocks

# ==================== å…­å¤§å®ˆè­·è€… ====================

def guardian_1_value_investing(stock, config):
    """å®ˆè­·è€… 1ï¼šåƒ¹å€¼æŠ•è³‡ï¼ˆå·´è²ç‰¹ï¼‰"""
    passed = []
    failed = []
    
    # æª¢æŸ¥ 1ï¼šæœ¬ç›Šæ¯”åˆç†
    if config["VALUE_MIN_PE"] <= stock["pe_ratio"] <= config["VALUE_MAX_PE"]:
        passed.append(f"âœ… PE={stock['pe_ratio']} åœ¨åˆç†ç¯„åœ")
    else:
        failed.append(f"âŒ PE={stock['pe_ratio']} ä¸åˆç†")
    
    # æª¢æŸ¥ 2ï¼šROE å¤ é«˜
    if stock["roe"] >= config["VALUE_MIN_ROE"]:
        passed.append(f"âœ… ROE={stock['roe']:.2%} å¤ é«˜")
    else:
        failed.append(f"âŒ ROE={stock['roe']:.2%} å¤ªä½")
    
    # æª¢æŸ¥ 3ï¼šåˆ©æ½¤ç‡å¥åº·
    if stock["profit_margin"] >= config["VALUE_MIN_PROFIT_MARGIN"]:
        passed.append(f"âœ… åˆ©æ½¤ç‡={stock['profit_margin']:.2%}")
    else:
        failed.append(f"âŒ åˆ©æ½¤ç‡={stock['profit_margin']:.2%} å¤ªä½")
    
    # æª¢æŸ¥ 4ï¼šè‚¡æ¯ç‡
    if stock["dividend_yield"] >= config["VALUE_MIN_DIV_YIELD"]:
        passed.append(f"âœ… è‚¡æ¯ç‡={stock['dividend_yield']:.2%}")
    else:
        failed.append(f"âŒ è‚¡æ¯ç‡={stock['dividend_yield']:.2%} å¤ªä½")
    
    is_pass = len(failed) == 0
    return {
        "guardian": "å®ˆè­·è€… 1ï¼šåƒ¹å€¼æŠ•è³‡",
        "passed": is_pass,
        "details": passed + failed,
        "score": len(passed) / (len(passed) + len(failed))
    }

def guardian_2_growth_investing(stock, config):
    """å®ˆè­·è€… 2ï¼šæˆé•·è‚¡æŠ•è³‡ï¼ˆå½¼å¾—æ—å€ï¼‰"""
    passed = []
    failed = []
    
    # æª¢æŸ¥ 1ï¼šç‡Ÿæ”¶æˆé•·ç‡
    if stock["revenue_growth"] >= config["GROWTH_MIN_REVENUE_GROWTH"]:
        passed.append(f"âœ… ç‡Ÿæ”¶æˆé•·={stock['revenue_growth']:.2%}")
    else:
        failed.append(f"âŒ ç‡Ÿæ”¶æˆé•·={stock['revenue_growth']:.2%} å¤ªæ…¢")
    
    # æª¢æŸ¥ 2ï¼šPEG æ¯”ç‡
    if stock["peg_ratio"] <= config["GROWTH_MAX_PEG"]:
        passed.append(f"âœ… PEG={stock['peg_ratio']} åˆç†")
    else:
        failed.append(f"âŒ PEG={stock['peg_ratio']} å¤ªé«˜")
    
    # æª¢æŸ¥ 3ï¼šEPS æˆé•·
    if stock["eps_growth"] >= config["GROWTH_MIN_PROFIT_GROWTH"]:
        passed.append(f"âœ… EPS æˆé•·={stock['eps_growth']:.2%}")
    else:
        failed.append(f"âŒ EPS æˆé•·={stock['eps_growth']:.2%} ä¸è¶³")
    
    is_pass = len(failed) == 0
    return {
        "guardian": "å®ˆè­·è€… 2ï¼šæˆé•·æŠ•è³‡",
        "passed": is_pass,
        "details": passed + failed,
        "score": len(passed) / (len(passed) + len(failed))
    }

def guardian_3_technical_analysis(stock, config):
    """å®ˆè­·è€… 3ï¼šæŠ€è¡“åˆ†æï¼ˆå¨å»‰æ­å°¼çˆ¾ï¼‰"""
    passed = []
    failed = []
    
    # æª¢æŸ¥ 1ï¼šç›¸å°å¼·åº¦
    if stock["rs_rank"] >= config["TECH_MIN_RS_RANK"]:
        passed.append(f"âœ… RS æ’å={stock['rs_rank']}")
    else:
        failed.append(f"âŒ RS æ’å={stock['rs_rank']} å¤ªä½")
    
    # æª¢æŸ¥ 2ï¼šæˆäº¤é‡æ”¾å¤§
    if stock["volume_ratio"] >= config["TECH_MIN_VOLUME_RATIO"]:
        passed.append(f"âœ… æˆäº¤é‡æ¯”={stock['volume_ratio']}")
    else:
        failed.append(f"âŒ æˆäº¤é‡æ¯”={stock['volume_ratio']} ä¸è¶³")
    
    # æª¢æŸ¥ 3ï¼šåƒ¹æ ¼ä½ç½®
    if stock["price_above_ma50"]:
        passed.append("âœ… åƒ¹æ ¼ç«™ä¸Š 50MA")
    else:
        failed.append("âŒ åƒ¹æ ¼ä½æ–¼ 50MA")
    
    is_pass = len(failed) == 0
    return {
        "guardian": "å®ˆè­·è€… 3ï¼šæŠ€è¡“åˆ†æ",
        "passed": is_pass,
        "details": passed + failed,
        "score": len(passed) / (len(passed) + len(failed))
    }

def guardian_4_risk_management(stock, config):
    """å®ˆè­·è€… 4ï¼šé¢¨éšªç®¡ç†ï¼ˆé›·é”é‡Œæ­ï¼‰"""
    passed = []
    failed = []
    
    # æª¢æŸ¥ 1ï¼šè² å‚µæ¯”ç‡
    if stock["debt_ratio"] <= config["RISK_MAX_DEBT_RATIO"]:
        passed.append(f"âœ… è² å‚µæ¯”={stock['debt_ratio']:.2%}")
    else:
        failed.append(f"âŒ è² å‚µæ¯”={stock['debt_ratio']:.2%} å¤ªé«˜")
    
    # æª¢æŸ¥ 2ï¼šæµå‹•æ¯”ç‡
    if stock["current_ratio"] >= config["RISK_MIN_CURRENT_RATIO"]:
        passed.append(f"âœ… æµå‹•æ¯”={stock['current_ratio']}")
    else:
        failed.append(f"âŒ æµå‹•æ¯”={stock['current_ratio']} å¤ªä½")
    
    # æª¢æŸ¥ 3ï¼šBeta å€¼
    if stock["beta"] <= config["RISK_MAX_BETA"]:
        passed.append(f"âœ… Beta={stock['beta']}")
    else:
        failed.append(f"âŒ Beta={stock['beta']} æ³¢å‹•å¤ªå¤§")
    
    is_pass = len(failed) == 0
    return {
        "guardian": "å®ˆè­·è€… 4ï¼šé¢¨éšªç®¡ç†",
        "passed": is_pass,
        "details": passed + failed,
        "score": len(passed) / (len(passed) + len(failed))
    }

def guardian_5_market_sentiment(stock, config):
    """å®ˆè­·è€… 5ï¼šå¸‚å ´æƒ…ç·’ï¼ˆç´¢ç¾…æ–¯ï¼‰"""
    passed = []
    failed = []
    
    # æª¢æŸ¥ 1ï¼šæ³•äººè²·ç›¤
    if stock["institutional_buy_ratio"] >= config["SENTIMENT_MIN_INST_BUY"]:
        passed.append(f"âœ… æ³•äººè²·ç›¤={stock['institutional_buy_ratio']:.2%}")
    else:
        failed.append(f"âŒ æ³•äººè²·ç›¤={stock['institutional_buy_ratio']:.2%} ä¸è¶³")
    
    # æª¢æŸ¥ 2ï¼šæ”¾ç©ºæ¯”ä¾‹
    if stock["short_interest_ratio"] <= config["SENTIMENT_MAX_SHORT_RATIO"]:
        passed.append(f"âœ… æ”¾ç©ºæ¯”={stock['short_interest_ratio']:.2%}")
    else:
        failed.append(f"âŒ æ”¾ç©ºæ¯”={stock['short_interest_ratio']:.2%} å¤ªé«˜")
    
    # æª¢æŸ¥ 3ï¼šåˆ†æå¸«è©•ç´š
    if stock["analyst_rating"] >= 3.5:
        passed.append(f"âœ… åˆ†æå¸«è©•åˆ†={stock['analyst_rating']}")
    else:
        failed.append(f"âŒ åˆ†æå¸«è©•åˆ†={stock['analyst_rating']} å¤ªä½")
    
    is_pass = len(failed) == 0
    return {
        "guardian": "å®ˆè­·è€… 5ï¼šå¸‚å ´æƒ…ç·’",
        "passed": is_pass,
        "details": passed + failed,
        "score": len(passed) / (len(passed) + len(failed))
    }

def guardian_6_macro_economics(stock, config):
    """å®ˆè­·è€… 6ï¼šç¸½é«”ç¶“æ¿Ÿï¼ˆé”é‡Œæ­ï¼‰"""
    passed = []
    failed = []
    
    # æ¨¡æ“¬ç¸½é«”ç¶“æ¿ŸæŒ‡æ¨™
    gdp_growth = random.uniform(0, 0.05)
    interest_rate = random.uniform(0.02, 0.06)
    
    # æª¢æŸ¥ 1ï¼šGDP æˆé•·ç‡
    if gdp_growth >= config["MACRO_MIN_GDP_GROWTH"]:
        passed.append(f"âœ… GDP æˆé•·={gdp_growth:.2%}")
    else:
        failed.append(f"âŒ GDP æˆé•·={gdp_growth:.2%} å¤ªä½")
    
    # æª¢æŸ¥ 2ï¼šåˆ©ç‡ç’°å¢ƒ
    if interest_rate <= config["MACRO_MAX_INTEREST_RATE"]:
        passed.append(f"âœ… åˆ©ç‡={interest_rate:.2%}")
    else:
        failed.append(f"âŒ åˆ©ç‡={interest_rate:.2%} å¤ªé«˜")
    
    # æª¢æŸ¥ 3ï¼šç”¢æ¥­æ™¯æ°£
    sector_score = random.uniform(0.5, 1.0)
    if sector_score >= 0.7:
        passed.append(f"âœ… ç”¢æ¥­æ™¯æ°£åˆ†æ•¸={sector_score:.2f}")
    else:
        failed.append(f"âŒ ç”¢æ¥­æ™¯æ°£åˆ†æ•¸={sector_score:.2f} åå¼±")
    
    is_pass = len(failed) == 0
    return {
        "guardian": "å®ˆè­·è€… 6ï¼šç¸½é«”ç¶“æ¿Ÿ",
        "passed": is_pass,
        "details": passed + failed,
        "score": len(passed) / (len(passed) + len(failed))
    }

# ==================== ä¸»åˆ†æå¼•æ“ ====================

def analyze_stock(stock, config):
    """å°å–®ä¸€è‚¡ç¥¨åŸ·è¡Œå…­å¤§å®ˆè­·è€…åˆ†æ"""
    results = [
        guardian_1_value_investing(stock, config),
        guardian_2_growth_investing(stock, config),
        guardian_3_technical_analysis(stock, config),
        guardian_4_risk_management(stock, config),
        guardian_5_market_sentiment(stock, config),
        guardian_6_macro_economics(stock, config),
    ]
    
    # è¨ˆç®—ç¸½åˆ†
    total_score = sum(r["score"] for r in results) / len(results)
    passed_count = sum(1 for r in results if r["passed"])
    
    return {
        "ticker": stock["ticker"],
        "name": stock["name"],
        "price": stock["price"],
        "guardian_results": results,
        "total_score": total_score,
        "passed_count": passed_count,
        "recommendation": "âœ… å¼·çƒˆæ¨è–¦" if passed_count >= 5 else
                         "âš ï¸ è¬¹æ…è€ƒæ…®" if passed_count >= 3 else
                         "âŒ ä¸å»ºè­°"
    }

def run_full_analysis(num_stocks=50, market_trend="bull", config=None):
    """åŸ·è¡Œå®Œæ•´åˆ†ææµç¨‹"""
    print(f"\n{'='*60}")
    print(f"ğŸ¯ Manpan æƒ…å ±ç¶² v4.1 - å…­å¤§å®ˆè­·è€…åˆ†æ")
    print(f"{'='*60}\n")
    
    # 1. ç”Ÿæˆè‚¡ç¥¨è³‡æ–™
    print(f"ğŸ“Š ç”Ÿæˆ {num_stocks} æª”æ¨¡æ“¬è‚¡ç¥¨ï¼ˆ{market_trend} å¸‚å ´ï¼‰...")
    stocks = generate_mock_stocks(num_stocks, market_trend)
    print(f"âœ… å®Œæˆ\n")
    
    # 2. åŸ·è¡Œåˆ†æ
    print("ğŸ” åŸ·è¡Œå…­å¤§å®ˆè­·è€…åˆ†æ...")
    results = []
    for stock in stocks:
        result = analyze_stock(stock, config or CONFIG)
        results.append(result)
    print(f"âœ… å®Œæˆ\n")
    
    # 3. æ’åºèˆ‡ç¯©é¸
    results.sort(key=lambda x: (x["passed_count"], x["total_score"]), reverse=True)
    
    # 4. è¼¸å‡ºå ±å‘Š
    print(f"\n{'='*60}")
    print(f"ğŸ“ˆ TOP 10 æ¨è–¦è‚¡ç¥¨")
    print(f"{'='*60}\n")
    
    for i, result in enumerate(results[:10], 1):
        print(f"\n{i}. {result['ticker']} - {result['name']}")
        print(f"   åƒ¹æ ¼ï¼š${result['price']:.2f}")
        print(f"   é€šéå®ˆè­·è€…ï¼š{result['passed_count']}/6")
        print(f"   ç¸½åˆ†ï¼š{result['total_score']:.2%}")
        print(f"   å»ºè­°ï¼š{result['recommendation']}")
        print(f"\n   å®ˆè­·è€…è©³æƒ…ï¼š")
        for guardian in result["guardian_results"]:
            status = "âœ…" if guardian["passed"] else "âŒ"
            print(f"   {status} {guardian['guardian']} ({guardian['score']:.1%})")
    
    # 5. çµ±è¨ˆæ‘˜è¦
    strong_buy = sum(1 for r in results if r["passed_count"] >= 5)
    consider = sum(1 for r in results if 3 <= r["passed_count"] < 5)
    avoid = sum(1 for r in results if r["passed_count"] < 3)
    
    print(f"\n\n{'='*60}")
    print(f"ğŸ“Š æ•´é«”çµ±è¨ˆ")
    print(f"{'='*60}")
    print(f"âœ… å¼·çƒˆæ¨è–¦ï¼š{strong_buy} æª”")
    print(f"âš ï¸ è¬¹æ…è€ƒæ…®ï¼š{consider} æª”")
    print(f"âŒ ä¸å»ºè­°ï¼š{avoid} æª”")
    print(f"\nå¸‚å ´é€šéç‡ï¼š{strong_buy/num_stocks:.1%}")
    print(f"{'='*60}\n")
    
    return results

print("âœ… ä¸»ç¨‹å¼è¼‰å…¥å®Œæˆ")
```
